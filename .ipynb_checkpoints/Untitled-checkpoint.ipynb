### Transform data from Wiki to pandas dataframe

```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame()
url = ['https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M']
for url in url:
    data = data.append(pd.read_html(url))
data.head()
```

```python
data = data[['Postcode', 'Borough', 'Neighbourhood']]
data.head()
print(data.shape)
```

```python
data
```

```python
data = data.drop(data[(data.Borough == "Not assigned")].index)
```

```python
data.head()
```

```python
data['Neighbourhood'].replace('Not assigned', data.Borough, inplace=True)
```

```python
temp_df = data.groupby('Postcode')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x))
temp_df = temp_df.reset_index(drop=False)
```

```python
temp_df
```

```python
temp_df.rename(columns={'Neighbourhood':'Neighbourhood_joined'},inplace=True)
```

```python
temp_df
```

```python
data_merge = pd.merge(data, temp_df, on='Postcode')
data_merge.drop(['Neighbourhood'],axis=1,inplace=True)
data_merge.drop_duplicates(inplace=True)
data_merge.rename(columns={'Neighbourhood_joined':'Neighbourhood'},inplace=True)
data_merge
```
